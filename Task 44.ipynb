```python
import pandas as pd
```

```python
import random
```

```python
lst = ['robot'] * 10
```

```python
lst += ['human'] * 10
```

```python
random.shuffle(lst)
```

```python
data = pd.DataFrame({'whoAmI': lst})
```

```python
data.head()
```

```python
data['tmp'] = 1
```

```python
data.set_index([data.index, 'whoAmI'], inplace=True)
```

```python
data = data.unstack(level=-1, fill_value = 0).astype(int)
```

```python
data.columns = data.columns.droplevel()
```

```python
data.columns.name = None
```
